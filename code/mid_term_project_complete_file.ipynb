## Adding categorical columns

```python
import pandas as pd
import os #provides functions for interacting with the operating system
import numpy as np 
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('concatenated_data.csv')
```

```python
data.tail()
```

```python
data.drop(['Unnamed: 0'], axis=1)
```

```python
data['year'] = data['year'].astype(int)
```

```python
# to change the year into date time format
data['year'] = data['year'].astype(str) + '-01-01'
data['year'] = pd.to_datetime(data['year'])
```

```python
data.head()
```

```python
data.age = pd.to_numeric(data.age, errors='coerce')
```

```python
data.drop(['Unnamed: 0'], axis=1)
```

```python
def cohort_definer(row):
    if row['age'] < 33:
        val = 'young'
    elif row['age'] >= 34 and row['age'] < 64:
        val = 'adult'
    else:
        val = 'senior'
    return val
data['cohort'] = data.apply(cohort_definer, axis=1)
```

```python
data.trust_church = pd.to_numeric(data.trust_church, errors='coerce')
```

```python
def trust_notrust(row):
    if row['trust_church'] == 2 or row['trust_church'] == 1:
        val = 'trust'
    elif row['trust_church'] == 3 or row['trust_church'] == 4:
        val = 'no_trust'
    else:
        val = 0
    return val
data['trust_notrust'] = data.apply(trust_notrust, axis=1)
```

```python
def gender_definer(row):
    if row['gender'] == 2:
        val = 'female'
    elif row['gender'] == 1:
        val = 'male'
    else:
        val = 'n/a'
    return val
data['gender_cat'] = data.apply(gender_definer, axis=1)
```

```python
def ec_cat(row):
    if row['economic_status'] == 1:
        val = 'very good'
    elif row['economic_status'] == 2:
        val = 'good'
    elif row['economic_status'] == 3:
        val = 'not bad'
    elif row['economic_status'] == 4:
        val = 'bad' 
    elif row['economic_status'] == 5:
        val = 'very bad'    
    else:
        val = 0
    return val
data['economic_status_cat'] = data.apply(ec_cat, axis=1)
```

```python
data.head()
```

```python
def judiciary(row):
    if row['economic_status'] == 1:
        val = 'very good'
    elif row['economic_status'] == 2:
        val = 'good'
    elif row['economic_status'] == 3:
        val = 'not bad'
    elif row['economic_status'] == 4:
        val = 'bad' 
    elif row['economic_status'] == 5:
        val = 'very bad'    
    else:
        val = 0
    return val
data['trust_judiciary_cat'] = data.apply(judiciary, axis=1)
```
